## second page
#### Suppose a store has three types of customers:  
##### Regular customers (60%)  
##### Premium customers (30%)  
##### VIP customers (10%)  

#### The probability that a customer makes a purchase given their category is:  
##### Regular: 0.4  
##### Premium: 0.7  
##### VIP: 0.9  
### What is the overall probability that a randomly chosen customer makes a purchase?

#### Answer:

$$ P[Buy == 1] = P[Buy|Regular] * P[Regular] + P[Buy|Premium]  * P[Premium] + P[Buy|VIP] * P[VIP] $$

```python
P_R = 0.60
P_P = 0.30
P_V = 0.10
P_B_R = 0.4
P_B_P = 0.7
P_B_V = 0.9

P_B = (P_B_R * P_R) + (P_B_P * P_P) + (P_B_V * P_V)
print(f"{round(P_B*100,2)}%")
```

## Third page

#### A bank is assessing whether a loan applicant will default on their loan. Based on historical data:  
##### 5% of all loan applicants default on their loans.  
##### 80% of those who defaulted had a low credit score.  
##### 30% of all applicants have a low credit score.  

### If a randomly chosen applicant has a low credit score, what is the probability that they will default?

#### Answer:

$$ P[default == 1 ] = 0.05 $$
$$ P[LowCreditScore| default] = 0.8 $$
$$ P[LowCreditScore] = 0.3 $$
$$ P[default|LowCreditScore] = ? $$ 
  
Base Law : 
$$ P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)} $$
so : 
$$ P[default|LowCreditScore] =  \frac{P[LowCreditScore| default] \cdot  P[default]}{P[LowCreditScore]} = \frac{0.8 \cdot  0.05}{0.3}   $$



```python
P_D = 0.05  
P_LS = 0.30  
P_LS_D = 0.80 

P_D_LS = (P_LS_D * P_D) / P_LS
print(f"{round(P_D_LS*100,2)}%")
```

#### Simulate a $10*1000$ matrix with $x_1, x_2, . . . , x_10 ∼ exp(2)  $  
#### Calculate a biased and unbiased estimator for $ V ar(X) =  \frac {1  } {λ^2} $  
#### Average on the results for both methods.  
#### Print the results.  
#### Write an expression for the difference.  

#### Answer:

	1.	Population Variance (Biased Estimator)

$$ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (X_i - \bar{X})^2$$

where  N  is the sample size.

	2.	Sample Variance (Unbiased Estimator)

$$ S^2 = \frac{1}{N-1} \sum_{i=1}^{N} (X_i - \bar{X})^2$$ 

    3.	Difference
$$ \frac{N}{N - 1} $$ 

```python
n = 10
matrix = np.random.exponential(scale=1/2, size=(n, 1000))
Biased = np.sum((matrix - np.mean(matrix)) ** 2 ) / n
unBiased = np.sum((matrix - np.mean(matrix)) ** 2 ) / ( n - 1)
mean_Biased = np.mean(Biased)
mean_unBiased = np.mean(unBiased)
print(f"Mean of Based: {mean_Biased}")
print(f"Mean of UnBiased: {mean_unBiased}")

difference = mean_unBiased / mean_Biased
difference_Theoretical = n / (n - 1)
print(f"Difference: {round(difference,3)}")
print(f"Theoretical Difference: {round(difference_Theoretical,3)}")
```

## Fifth page

#### Simulate 1000 samples of size 100 from the exp(3) distribution.  
### What is the mean E(X) of exp(3)? What is the variance and sd?  



#### Answer:




$$ E[X] = \frac{1}{\lambda} $$

$$ \hat{E[X]} = \bar{X} $$

$$ \text{Var}(X) = \frac{1}{\lambda^2} $$ 

$$ \hat{\text{Var}}(X) = \frac{1}{\hat{\lambda}^2} = \bar{X}^2 $$ 

$$ \text{SD}(X) = \sqrt{\text{Var}(X)} = \frac{1}{\lambda} $$


$$ \hat{\lambda} = \frac{1}{\bar{X}} $$

$$ \hat{\text{SD}}(X) = \frac{1}{\hat{\lambda}} =  \bar{X} $$ 





```python
n = 1000
sample_size = 100  # גודל כל דגימה
lambda_val = 3
samples = np.random.exponential(scale=1/lambda_val, size=(n, sample_size))

true_mean = 1 / lambda_val

sample_means = np.mean(samples, axis=1)
mean_of_samples = np.mean(sample_means, axis=0)
print(f"True Mean: {  round(true_mean,4)}")
print(f"Mean of Samples: {  round(mean_of_samples,4) }")
print()
true_variance = 1 / (lambda_val ** 2)
var_samples = mean_of_samples ** 2
print(f"True Variance: {  round(true_variance,4)}")
print(f"Variance from Samples: {  round(var_samples,4)}")
print()
true_sd = np.sqrt(true_variance)
print(f"True Standard Deviation: { round(true_sd,4)}")
sample_sd = mean_of_samples
print(f"Standard Deviation from Samples: {round(sample_sd,4)}")
```

### Test the following two sided hypotheses:
$$H0 : µ0 = E[exp(3)]$$
$$ H1 : µ0  \neq E[exp(3)] $$

### Use both methods learned in class.  
### Calculate the proportion of times that H0 was rejected.  

#### Answer:

##### **ביצוע בדיקת השערות באמצעות ערך קריטי**

נבצע בדיקה על כל הדגימות, ונבדוק בכל פעם האם ממוצע הדגימה חורג מהערך הקריטי \($c$ \). אם כן, נדחה את \($H_0$\) י .

##### **ביצוע בדיקת השערות באמצעות $pvalue$  י**

בבדיקה זו, נחשב עבור כל דגימה את **סטטיסטי הבדיקה ($Z$)** ואת **$p-value$** המתאים. אם ה-$p-value$ קטן מרמת המובהקות \($\alpha$\), נדחה את \($H_0$\) י.

```python
std = np.std(sample_means, ddof=1)   # סטיית תקן 
Z = (sample_means - true_mean) / std # Z ערך

# ערך קריטי
Z_critical = stats.t.ppf( 0.975 , df=sample_size - 1) #ערך קריטי
reject_h0_critical = np.abs(Z) > Z_critical
print(f"{round(np.mean(reject_h0_critical)*100,3)}%")

# Pvalue
p_value = (1 - stats.t.cdf(np.abs(Z), df=sample_size - 1))
reject_h0_pval = p_value < 0.05/2
print(f"{round(np.mean(reject_h0_pval)*100,3)}%")
```

### Write the pmf if given $Y , X_1, . . . , X_n $ are independent.  


#### Answer:
$$
P_{Y, X_1, X_2, \dots, X_n}(y, x_1, x_2, \dots, x_n) = P_Y(y) \cdot P_{X_1}(x_1) \cdot P_{X_2}(x_2) \dots \space \cdot P_{X_n}(x_n)
$$ 

### Write the density of two independent normal variables.  


#### Answer:
Density of $X , Y $ if $X , Y $ are normal: 
$$
f X (x) = \frac{1}{\sqrt{2\pi \sigma_x^2}} \exp \left( -\frac{(x - \mu)^2}{2\sigma_x^2} \right)
$$
$$
f Y (y) = \frac{1}{\sqrt{2\pi \sigma_y^2}} \exp \left( -\frac{(y - \mu)^2}{2\sigma_y^2} \right)
$$

Joint density:
$$
f_{X,Y}(x, y) = f_X(x) \cdot f_Y(y)  = \frac{1}{2\pi \sigma_x \sigma_y} \exp \left( -\frac{(x - \mu_x)^2}{2\sigma_x^2} - \frac{(y - \mu_y)^2}{2\sigma_y^2} \right)
$$
